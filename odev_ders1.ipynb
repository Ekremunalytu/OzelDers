## Bolum 2: Matematik Islemleri

Python bir hesap makinesi gibi calisir:

| Islem | Sembol | Ornek | Sonuc |
|-------|--------|-------|-------|
| Toplama | `+` | `10 + 3` | `13` |
| Cikarma | `-` | `10 - 3` | `7` |
| Carpma | `*` | `10 * 3` | `30` |
| Bolme | `/` | `10 / 3` | `3.33` |
| Tam bolme | `//` | `10 // 3` | `3` |
| Kalan (mod) | `%` | `10 % 3` | `1` |
| Us alma | `**` | `10 ** 3` | `1000` |

````python
# ORNEK: Indirim hesabi
fiyat = 150
indirim_yuzde = 20
indirim_tutari = fiyat * indirim_yuzde / 100
son_fiyat = fiyat - indirim_tutari

print(f"Fiyat: {fiyat} TL")
print(f"%{indirim_yuzde} indirimle: {son_fiyat} TL")
````

## Bolum 3: if / elif / else

Bilgisayara kosula gore karar verdiriyoruz:

```
  Soru: Not 50'den buyuk mu?
          |
      +---+---+
      |       |
    EVET    HAYIR
      |       |
   Gecti    Kaldi
```

**DIKKAT:** `if`'ten sonra 4 bosluk (girinti) SART!

````python
# ORNEK: Sicakliga gore oneri
sicaklik = 35

if sicaklik >= 35:
    print(f"{sicaklik} derece - Cok sicak!")
elif sicaklik >= 25:
    print(f"{sicaklik} derece - Guzel hava.")
elif sicaklik >= 15:
    print(f"{sicaklik} derece - Serin, mont al.")
else:
    print(f"{sicaklik} derece - Soguk!")
````

## Bolum 4: Listeler

Liste = Sirali cekmece. Her gozun bir numarasi var (0'dan baslar!).

```
  notlar = [80, 55, 90, 70]

  +----+----+----+----+
  | 80 | 55 | 90 | 70 |
  +----+----+----+----+
   [0]  [1]  [2]  [3]
```

| Komut | Ne Yapar |
|-------|----------|
| `liste[0]` | Ilk elemani getirir |
| `liste[-1]` | Son elemani getirir |
| `liste.append(x)` | Sona x ekler |
| `liste.remove(x)` | x degerini cikarir |
| `len(liste)` | Eleman sayisini verir |

````python
# ORNEK: Liste islemleri
renkler = ["kirmizi", "mavi", "yesil"]
print("Liste:", renkler)
print("Ilk renk:", renkler[0])
print("Son renk:", renkler[-1])

renkler.append("sari")
print("Sari eklendi:", renkler)
print("Eleman sayisi:", len(renkler))
````
