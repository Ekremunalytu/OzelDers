## BOLUM 3: Fonksiyon (def)
---

Aynı kodu tekrar tekrar yazmamak için **fonksiyon** kullanırız.

Fonksiyon = bir işi yapan **küçük makine**. Butona basarsın, işi yapar.

### 3.1 — İlk fonksiyonumuz

````python
def selamla():
    print("Merhaba!")
    print("Derse hoş geldin!")

# Fonksiyonu çağır (butona bas)
selamla()
````

### 3.2 — Parametre alan fonksiyon

Fonksiyona dışarıdan bilgi gönderebilirsin.

````python
def selamla(isim):
    print(f"Merhaba {isim}!")

selamla("Ali")
selamla("Ayşe")
selamla("Mehmet")
````

### 3.3 — Değer döndüren fonksiyon (return)

````python
def topla(a, b):
    sonuc = a + b
    return sonuc

x = topla(3, 5)
print(f"3 + 5 = {x}")

y = topla(10, 20)
print(f"10 + 20 = {y}")
````

### 3.4 — Fonksiyon + if birlikte

````python
def not_kontrol(not_):
    if not_ >= 50:
        return "Gecti"
    else:
        return "Kaldi"

sonuc = not_kontrol(75)
print(sonuc)

sonuc = not_kontrol(40)
print(sonuc)
````

## BOLUM 4: Class ve Object — İlk Adım
---

Şimdiye kadar verileri ayrı ayrı değişkenlerde tuttuk. Ama bir öğrencinin hem adı, hem numarası, hem notları var. Bunları **bir arada** tutmak için **class** kullanırız.

**Class** = şablon, kalıp
**Object** = o şablondan yapılmış gerçek şey

Örnek:
- Class = Öğrenci kartı şablonu (isim yeri var, numara yeri var)
- Object = Ali'nin kartı, Ayşe'nin kartı (gerçek bilgiler doldurulmuş)

### 4.1 — En basit class

````python
class Ogrenci:
    def __init__(self, isim, numara):
        self.isim = isim
        self.numara = numara

# __init__ = nesne oluşturulunca otomatik çalışır
# self = "bu nesnenin kendisi" demek

print("Class tanımlandı! Henüz nesne oluşturmadık.")
````

### 4.2 — Class'tan nesne (object) oluşturma

````python
# Nesne oluştur
ali = Ogrenci("Ali", 101)
ayse = Ogrenci("Ayşe", 102)

# Bilgilerine eriş (nokta ile)
print(ali.isim)
print(ali.numara)

print(ayse.isim)
print(ayse.numara)
````

## BOLUM 5: Class'a Method Ekleme
---

Method = class'ın içindeki fonksiyon. Nesnenin **yapabileceği iş**.

Düşün: telefon bir class. Arama yapmak, mesaj göndermek → methodlar.

### 5.1 — İlk method: kendini tanıt

````python
class Ogrenci:
    def __init__(self, isim, numara):
        self.isim = isim
        self.numara = numara

    def tanitim(self):
        print(f"Ben {self.isim}, numaram {self.numara}.")

# Nesne oluştur ve methodu çağır
ali = Ogrenci("Ali", 101)
ali.tanitim()

ayse = Ogrenci("Ayşe", 102)
ayse.tanitim()
````

### 5.2 — Liste tutan class + not ekleme methodu

````python
class Ogrenci:
    def __init__(self, isim, numara):
        self.isim = isim
        self.numara = numara
        self.notlar = []          # boş liste — notlar sonra eklenecek

    def tanitim(self):
        print(f"Ben {self.isim}, numaram {self.numara}.")

    def not_ekle(self, yeni_not):
        self.notlar.append(yeni_not)
        print(f"{self.isim}'e {yeni_not} notu eklendi.")

# Kullanalım
ali = Ogrenci("Ali", 101)
ali.not_ekle(85)
ali.not_ekle(90)
ali.not_ekle(72)

print(f"Ali'nin notları: {ali.notlar}")
````

### 5.3 — Ortalama hesaplayan method

````python
class Ogrenci:
    def __init__(self, isim, numara):
        self.isim = isim
        self.numara = numara
        self.notlar = []

    def not_ekle(self, yeni_not):
        self.notlar.append(yeni_not)

    def ortalama(self):
        toplam = 0
        for n in self.notlar:
            toplam = toplam + n
        return toplam / len(self.notlar)

# Kullanalım
ali = Ogrenci("Ali", 101)
ali.not_ekle(85)
ali.not_ekle(90)
ali.not_ekle(72)

ort = ali.ortalama()
print(f"{ali.isim}'in ortalaması: {ort}")
````

## BOLUM 6: Kontrol (Validation) Ekleme
---

Şimdi bir sorun var: birisi -50 veya 200 gibi saçma bir not girerse ne olur?

Bunu engellemek için method'un içine **if kontrolü** ekliyoruz.

````python
class Ogrenci:
    def __init__(self, isim, numara):
        self.isim = isim
        self.numara = numara
        self.notlar = []

    def not_ekle(self, yeni_not):
        if yeni_not < 0 or yeni_not > 100:
            print(f"HATA: {yeni_not} geçerli değil! (0-100 arası gir)")
        else:
            self.notlar.append(yeni_not)
            print(f"{self.isim}'e {yeni_not} eklendi.")

    def ortalama(self):
        if len(self.notlar) == 0:
            print("Henüz not yok!")
            return 0
        toplam = 0
        for n in self.notlar:
            toplam = toplam + n
        return toplam / len(self.notlar)

# Deneyelim
ali = Ogrenci("Ali", 101)
ali.not_ekle(85)
ali.not_ekle(-10)    # HATA verecek
ali.not_ekle(200)    # HATA verecek
ali.not_ekle(90)

print(f"Notlar: {ali.notlar}")
print(f"Ortalama: {ali.ortalama()}")
````

Gördüğün gibi -10 ve 200 listeye **eklenmedi**. Kontrol çalıştı!

## BOLUM 7: Nesneleri Liste İçinde Tutma
---

Birden fazla öğrenciyi tek tek değişkenlerde tutmak yerine, **hepsini bir listeye** koyabiliriz.

Sonra döngü ile hepsinin üzerinde işlem yaparız.

````python
# Öğrencileri oluştur
ali = Ogrenci("Ali", 101)
ali.not_ekle(85)
ali.not_ekle(90)

ayse = Ogrenci("Ayşe", 102)
ayse.not_ekle(92)
ayse.not_ekle(88)

mehmet = Ogrenci("Mehmet", 103)
mehmet.not_ekle(45)
mehmet.not_ekle(52)

# Hepsini bir listeye koy
sinif = [ali, ayse, mehmet]
````

````python
# Döngü ile hepsini gez
for ogr in sinif:
    ort = ogr.ortalama()
    print(f"{ogr.isim} - Ortalama: {ort}")
````

````python
# Kaç kişi geçti, kaç kişi kaldı?
gecen = 0
kalan = 0

for ogr in sinif:
    if ogr.ortalama() >= 50:
        gecen = gecen + 1
    else:
        kalan = kalan + 1

print(f"Geçen: {gecen}")
print(f"Kalan: {kalan}")
````

## BOLUM 8: Mini Proje — Sınıf Karnesi
---

Şimdi öğrendiğimiz her şeyi birleştiriyoruz:
- class + list + method + if kontrolü + for döngüsü

**Senaryo:** Öğrencilerin notlarını gir, karnelerini oluştur, en başarılı öğrenciyi bul.

````python
class Ogrenci:
    def __init__(self, isim, numara):
        self.isim = isim
        self.numara = numara
        self.notlar = []

    def not_ekle(self, yeni_not):
        if yeni_not < 0 or yeni_not > 100:
            print(f"HATA: {yeni_not} gecerli degil!")
        else:
            self.notlar.append(yeni_not)

    def ortalama(self):
        if len(self.notlar) == 0:
            return 0
        toplam = 0
        for n in self.notlar:
            toplam = toplam + n
        return toplam / len(self.notlar)

    def durum(self):
        if self.ortalama() >= 50:
            return "Gecti"
        else:
            return "Kaldi"
````

````python
# Öğrencileri oluştur ve notlarını gir
sinif = []

o1 = Ogrenci("Ali", 101)
o1.not_ekle(85)
o1.not_ekle(90)
o1.not_ekle(78)
sinif.append(o1)

o2 = Ogrenci("Ayşe", 102)
o2.not_ekle(92)
o2.not_ekle(88)
o2.not_ekle(95)
sinif.append(o2)

o3 = Ogrenci("Mehmet", 103)
o3.not_ekle(45)
o3.not_ekle(52)
o3.not_ekle(38)
sinif.append(o3)

o4 = Ogrenci("Zeynep", 104)
o4.not_ekle(70)
o4.not_ekle(65)
o4.not_ekle(72)
sinif.append(o4)

print(f"Sınıfta {len(sinif)} öğrenci var.")
````

````python
# Karne oluştur
print("=" * 45)
print("SINIF KARNESİ")
print("=" * 45)

for ogr in sinif:
    ort = ogr.ortalama()
    print(f"{ogr.numara} - {ogr.isim:10s} | Ort: {ort:.1f} | {ogr.durum()}")

print("=" * 45)
````

````python
# En başarılı öğrenciyi bul
en_yuksek = 0
en_basarili = ""

for ogr in sinif:
    if ogr.ortalama() > en_yuksek:
        en_yuksek = ogr.ortalama()
        en_basarili = ogr.isim

print(f"En başarılı: {en_basarili} ({en_yuksek:.1f} ortalama)")
````

## BOLUM 9: Kotlin / Android Geçiş Hazırlığı
---

Tebrikler! Python'da **list, class, object, method, if kontrolü** kavramlarını öğrendin.

Bu kavramlar Kotlin'de de neredeyse aynı. Aşağıda karşılaştırma var — **şimdilik sadece oku, benzerlikleri bul.**

### Python vs Kotlin Tablosu

| Ne yapıyorsun? | Python | Kotlin |
|---|---|---|
| Yazdırma | `print("Merhaba")` | `println("Merhaba")` |
| Değişken | `isim = "Ali"` | `val isim = "Ali"` |
| Liste | `notlar = [85, 90]` | `val notlar = arrayListOf(85, 90)` |
| Ekleme | `notlar.append(75)` | `notlar.add(75)` |
| Döngü | `for x in liste:` | `for (x in liste) { }` |
| Koşul | `if x > 50:` | `if (x > 50) { }` |
| Fonksiyon | `def hesapla():` | `fun hesapla() { }` |
| Class | `class Ogrenci:` | `class Ogrenci { }` |
| Nesne | `ali = Ogrenci("Ali")` | `val ali = Ogrenci("Ali")` |

### Aynı Class — Kotlin'de nasıl görünür?

Sadece oku ve Python ile benzerliklerini bul:

```kotlin
class Ogrenci(val isim: String, val numara: Int) {
    val notlar = arrayListOf<Int>()

    fun notEkle(yeniNot: Int) {
        if (yeniNot < 0 || yeniNot > 100) {
            println("HATA: $yeniNot gecerli degil!")
        } else {
            notlar.add(yeniNot)
        }
    }

    fun ortalama(): Double {
        var toplam = 0
        for (n in notlar) {
            toplam += n
        }
        return toplam.toDouble() / notlar.size
    }
}

val ali = Ogrenci("Ali", 101)
ali.notEkle(85)
println(ali.ortalama())
```

### Android'de Akış

Android uygulamalarında aynı mantık var:

```
Kullanıcı veri girer (EditText)  →  input()
Butona basar (Button)            →  fonksiyon çağırma
Kod çalışır (class, method)      →  işlem
Sonuç gösterilir (TextView)      →  print()
```

Yani bu derste öğrendiğin her şey Android'in **arka planında** çalışan mantık!
