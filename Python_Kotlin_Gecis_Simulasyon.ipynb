## Bölüm 4: Method Kullanımı

Şimdi class içine sadece veri koymayacağız.
Aynı zamanda nesnenin yapabildiği işleri de ekleyeceğiz.

**Method = class içindeki fonksiyon**

Yani artık oyuncu sadece veri taşıyan bir şey değil, işlem yapabilen bir yapı olacak.

```python
class Oyuncu:
    def __init__(self, isim, takim):
        self.isim = isim
        self.takim = takim
        self.puanlar = []

    def tanit(self):
        print(f"Ben {self.isim}. Takımım: {self.takim}")

    def puan_ekle(self, yeni_puan):
        self.puanlar.append(yeni_puan)


ece = Oyuncu("Ece", "Mavi Takım")
ece.tanit()
ece.puan_ekle(90)
ece.puan_ekle(84)

print(ece.puanlar)
```

### Burada Yeni Olan Nedir?

- `tanit()` nesnenin kendini tanıtmasını sağlar.
- `puan_ekle(yeni_puan)` dışarıdan bilgi alır.
- `self.puanlar` nesnenin içinde saklanan listedir.

Yani methodlar sayesinde veri ile işlem aynı yapının içinde birleşmiş olur.

```python
class Oyuncu:
    def __init__(self, isim, takim):
        self.isim = isim
        self.takim = takim
        self.puanlar = []

    def puan_ekle(self, yeni_puan):
        self.puanlar.append(yeni_puan)

    def ortalama(self):
        toplam = 0
        for puan in self.puanlar:
            toplam = toplam + puan
        return toplam / len(self.puanlar)

    def en_yuksek_puan(self):
        en_buyuk = 0
        for puan in self.puanlar:
            if puan > en_buyuk:
                en_buyuk = puan
        return en_buyuk


mert = Oyuncu("Mert", "Kırmızı Takım")
mert.puan_ekle(76)
mert.puan_ekle(88)
mert.puan_ekle(81)

print("Ortalama:", mert.ortalama())
print("En yüksek puan:", mert.en_yuksek_puan())
```

Bu örnekte oyuncu için artık şunları yapabiliyoruz:

## Bölüm 5: Kontrol Mekanizması - Validation

Şimdi önemli bir problem var.

Bir kullanıcı yanlışlıkla şu değerleri girebilir:
- `-10`
- `150`
- `999`

Bu değerler mantıklı değilse sisteme eklenmemeli.
İşte buna **validation** deriz.

```python
class Oyuncu:
    def __init__(self, isim, takim):
        self.isim = isim
        self.takim = takim
        self.puanlar = []

    def puan_ekle(self, yeni_puan):
        if yeni_puan < 0 or yeni_puan > 100:
            print(f"HATA: {yeni_puan} geçerli değil.")
        else:
            self.puanlar.append(yeni_puan)
            print(f"{self.isim} için {yeni_puan} puanı eklendi.")

    def ortalama(self):
        if len(self.puanlar) == 0:
            return 0

        toplam = 0
        for puan in self.puanlar:
            toplam = toplam + puan
        return toplam / len(self.puanlar)

    def durum(self):
        if self.ortalama() >= 80:
            return "Yüksek performans"
        elif self.ortalama() >= 50:
            return "Orta performans"
        else:
            return "Geliştirilmeli"
```

```python
deniz = Oyuncu("Deniz", "Mavi Takım")

deniz.puan_ekle(85)
deniz.puan_ekle(-5)
deniz.puan_ekle(130)
deniz.puan_ekle(72)

print("Geçerli puanlar:", deniz.puanlar)
print("Ortalama:", deniz.ortalama())
print("Durum:", deniz.durum())
```

### Neden Önemli?

Gerçek hayatta kullanıcı hatalı veri girebilir.
İyi yazılmış programlar bu veriyi kontrol eder.

Öğrenci burada şunu anlamalı:

**Sadece işlem yapmak yetmez, veriyi korumak da gerekir.**

Bu düşünce ileride form kontrolü, Android input kontrolü ve backend doğrulama mantığı için temel oluşturur.

## Bölüm 6: Mini Proje - Oyun Kulübü Raporu

Şimdi öğrendiğimiz her şeyi tek bir akışta birleştireceğiz.

Bu bölümde:
- nesne oluşturacağız,
- nesneleri bir listede tutacağız,
- döngü ile hepsini gezeceğiz,
- ortalamaları hesaplayacağız,
- en başarılı oyuncuyu bulacağız.

### Algoritma Planı

Kod yazmadan önce adımları görmek önemlidir:

1. Boş bir `kulup` listesi oluştur.
2. Oyuncu nesneleri üret.
3. Her oyuncuya puan ekle.
4. Oyuncuları listeye koy.
5. Döngü ile oyuncuları sırayla gez.
6. Ortalama hesapla.
7. En yüksek ortalamayı takip et.
8. Sonucu ekrana yazdır.

Bu, algoritmik düşünmenin en temel örneklerinden biridir.

```python
kulup = []

oyuncu_a = Oyuncu("Ece", "Mavi Takım")
oyuncu_a.puan_ekle(90)
oyuncu_a.puan_ekle(85)
oyuncu_a.puan_ekle(93)
kulup.append(oyuncu_a)

oyuncu_b = Oyuncu("Mert", "Kırmızı Takım")
oyuncu_b.puan_ekle(70)
oyuncu_b.puan_ekle(74)
oyuncu_b.puan_ekle(68)
kulup.append(oyuncu_b)

oyuncu_c = Oyuncu("Deniz", "Mavi Takım")
oyuncu_c.puan_ekle(88)
oyuncu_c.puan_ekle(91)
oyuncu_c.puan_ekle(86)
kulup.append(oyuncu_c)

print("Kulüpteki oyuncu sayısı:", len(kulup))
```

```python
en_iyi_isim = ""
en_iyi_ortalama = 0
hedefi_gecen = 0

for oyuncu in kulup:
    ort = oyuncu.ortalama()
    print(f"{oyuncu.isim} | Takım: {oyuncu.takim} | Ortalama: {ort:.1f} | Durum: {oyuncu.durum()}")

    if ort >= 80:
        hedefi_gecen = hedefi_gecen + 1

    if ort > en_iyi_ortalama:
        en_iyi_ortalama = ort
        en_iyi_isim = oyuncu.isim

print("-" * 45)
print("80 ve üstü ortalamaya sahip oyuncu sayısı:", hedefi_gecen)
print(f"En başarılı oyuncu: {en_iyi_isim} ({en_iyi_ortalama:.1f})")
```

## Bölüm 8: Python -> Kotlin / Android Geçişi

Şimdi önemli bağlantıyı kuralım.

Ama kurduğumuz mantık Kotlin tarafında da devam eder.

| Python | Kotlin |
|---|---|
| `print()` | `println()` |
| `liste = []` | `val liste = arrayListOf()` |
| `append()` | `add()` |

| `class` | `class` |
| `self` | `this` mantığı |

- kullanıcı puan girer,
